```python
import pandas as pd
```

```python
data = pd.read_csv('performances.csv')
data.head()
```

```python
data['attendance'] = data.groupby(['group', 'round'])['choice'].transform('sum')
data.head()
```

```python
def find_score(line,threshold) :
    if line['choice'] == 1:
        if line['attendance'] <= threshold:
            return 1
        else:
            return -1
    else:
        return 0
```

```python
data['score'] = data.apply(lambda x: find_score(x), axis=1)
data.head()
```

```python
grupo1 = data[data['group'] == 'Grupo-0001']
grupo1.groupby('round')['choice'].sum().plot()
```

```python
grupo2 = data[data['group'] == 'Grupo-0002']
grupo2.groupby('round')['choice'].sum().plot()
```

```python
data['ac_score'] = data.groupby(['group', 'player'])['score'].transform('sum')
data.head()
```
